## Datos distribuidos

```python
# Importamos la librer√≠a
import pandas as pd
```

```python
# Cargamos el primer DataSet para tener con quien concatenar
data = pd.read_csv("../datasets/distributed-data/001.csv")
data.head()
```

```python
data.shape
```

* Importar el primer fichero.
* Hacemos un bucle para ir recorriendo todos y cada uno de los ficheros.
    * Importante tener una consistencia en el nombre de los ficheros.
    * Importamos los ficheros uno a uno.
    * Cada uno de ellos deben aprendizarse (a√±adir al final) del primer fichero que ya habiamos cargado
* Repetimos el bucle hasta que no queden ficheros

```python
filepath = "../datasets/distributed-data/"

data = pd.read_csv("../datasets/distributed-data/001.csv")

final_length = len(data)

for i in range(2, 333):
    if i < 10:
        filename = "00" + str(i)
    elif 10 <= i < 100:
        filename = "0" + str(i)
    elif i >= 100:
        filename = str(i)
    
    file = filepath + filename + ".csv"
    temp_data = pd.read_csv(file)
    
    final_length += len(temp_data)
    
    data = pd.concat([data, temp_data], axis=0)
```

```python
data.shape
```

```python
final_length
```

```python
final_length == data.shape[0]
```

## Tipo de Joins

```python
from IPython.display import Image
import numpy as np
```

**Inner Join <= A (Left Join), B (Right Join) <= Outer Join**

### Inner Join

* Devuelve un data frame con las filas que tienen valor tanto en el primer como 
  en el segundo data frame que estamos uniendo.
* El n√∫mero de filas ser√° igual al n√∫mero de filas comunes que tengan ambos sets.
    * Data Set A tiene 60 filas.
    * Data Set B tiene 50 filas.
    * Ambos comparten 30 filas.
    * Entonces A Inner Join B tendr√° 30 filas.
* En terminos de teor√≠a de conjuntos, se trata de la intersecci√≥n de los dos conjuntos.

```python
 # Atletas que se deben mantener
out_athletes = np.random.choice(data_main["Athlete"], size = 6, replace = False)
```

```python
# Se elimina todo menos los atletas de arriba y Michael Phelps
data_country_dlt = data_country_dp[(~data_country_dp["Athlete"].isin(out_athletes)) & 
                                  (data_country_dp["Athlete"] != "Michael Phelps")]

data_sports_dlt = data_sports_dp[(~data_sports_dp["Athlete"].isin(out_athletes)) & 
                                  (data_sports_dp["Athlete"] != "Michael Phelps")]

data_main_dlt = data_main[(~data_main["Athlete"].isin(out_athletes)) & 
                                  (data_main["Athlete"] != "Michael Phelps")]
```

```python
len(data_country_dlt)
```

```python
len(data_sports_dlt)
```

```python
len(data_main_dlt)
```

```python
data_country_dlt.head()
```

```python
Image(filename="../datasets/join/inner-join.png")
```

```python
# data_main contiene toda la info
# data_country_dlt le falta la info de 7 atletas
merged_inner = pd.merge(left=data_main, right=data_country_dlt,
                       how="inner", left_on="Athlete", right_on="Athlete")
```

```python
merged_inner.head()
```

### Left Join

* Devuelve un data frame con las filas que tuvieran valor en el dataset
  de la izquierda, sin importar si tienen correspondencia en el de la 
  derecha o no.
* Las filas del data frame final ser√° igual al n√∫mero de filas del data 
  frame izquierdo.
    * Data Set A tiene 60 filas.
    * Data Set B tiene 50 filas.
    * Entonces, A Left Join B tendr√° 60 filas.
* En t√©rminos teor√≠a de conjuntos, se trata del propio data set de izquierda
  quien, adem√°s tiene la intersercci√≥n en su interior.

```python
Image(filename="../datasets/join/left-join.png")
```

```python
merged_left = pd.merge(left=data_main, right=data_country_dlt, how="left", left_on="Athlete", right_on="Athlete")
```

```python
merged_left.head()
```

### Right Join

* Devuelve un data frame con las filas que tuvieran valor en el dataframe de la derecha, sin importar si tiene correspondencia en el izquierdo o no.
* Las filas del data frame final que no correspondan a ninguna fila del data frame izquierdo, tendr√°n NaN en las columnas del data frame izquierdo.
* El n√∫mero de filas ser√° igual al n√∫mero de filas del data frame derecho.
    * Data Set A tiene 60 filas.
    * Data Set B tiene 50 filas.
    * Entonces, A Right Join B tendr√° 50 filas.
* En terminos de teoria de conjuntos, se trata del propio data set de la derecha quien, adem√°s tiene la intersecci√≥n en su interior.    

```python
Image(filename="../datasets/join/right-join.png")
```

```python
merged_right = pd.merge(left=data_main_dlt, right=data_country_dp, how="right", left_on="Athlete", right_on="Athlete")
```

```python
merged_right.head()
```

### Outer Join

* Devuelve un data frame con todas las filas de ambos, reemplazando las ausencias de uno o de otro con NaN en la regi√≥n espec√≠fica.
* Las filas del data frame final que no correspondan a ninguna fila del data frame derecho (o izquierdo), tendr√°n NaN en las columnas del data frame derecho (o izquierdo).
* El n√∫mero de filas ser√° igual al m√°ximo n√∫mero de filas de ambos data frames.
    * Data Set A tiene 60 filas.
    * Data Set B tiene 50 filas.
    * Entonces, A Outer Join B tendr√° 60 + 50 = 80 filas.
* En t√©rminos de teor√≠a de conjuntos, se trata de la uni√≥n de conjuntos.        

```python
Image(filename="../datasets/join/outer-join.png")
```

```python
data_country_jb = data_country_dlt.append(
    {
        "Athlete" : "Juan Gabriel Gomila",
        "Country" : "Espa√±a"
    },ignore_index = True
)
```
