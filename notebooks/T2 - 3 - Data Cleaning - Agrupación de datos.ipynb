# Agregación de datos por categoría

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(2022)
```

```python
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
n = 500

gender_data = []
income_data = []

for i in range(0,n):
    # choice para eegir aleatorio de una lista
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
gender_data[0:10]
```

```python
income_data[0:10]
```

```python
# Z -> N(0,1)
# N(m,s) -> m + s * Z
height = np.ceil(170 + 20 * np.random.randn(n))
weight = np.ceil(80 + 30 * np.random.randn(n))
age = np.ceil(40 + 10 * np.random.randn(n))
income = np.ceil(18000 + 3500 * np.random.randn(n))
```

```python
data = pd.DataFrame(
    {
        "Gender" : gender_data,
        "Economic Status" : income_data,
        "height" : height,
        "Weight" : weight,
        "Age" : age,
        "income" : income
    }
)
```

```python
data
```

### Agrupación de datos

```python
# Se usa el groupby con alguna de las llaves del DataFrame
grouped_gender = data.groupby("Gender")
```

```python
# Mostramos la información
grouped_gender.groups
```

```python
# Recorremos y mostramos los grupos
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
# Otra forma de objener información
grouped_gender.get_group("Male")
```

```python
grouped_gender.get_group("Female")
```

```python
# Agrupamos por dos factores
double_group = data.groupby(["Gender", "Economic Status"])
```

```python
len(double_group)
```

```python
# Mostramos este grupo
for names, groups in double_group:
    print(names)
    print(groups)
```

### Operaciones sobre datos agrupados

```python
# Sumamos el grupo
double_group.sum()
```

```python
# Media del grupo
double_group.mean()
```

```python
# Cuantos elementos tiene cada categoria del grupo
double_group.size()
```

```python
# Obtenemos los estádisticos
double_group.describe()
```

```python
# Filtramos solo por ingreso
grouped_income = double_group["income"]
```

```python
# Estadisticas solo de income
grouped_income.describe()
```

```python
# Agregamos algunos estádisticos
double_group.aggregate(
    {
        "income" : np.sum,
        "Age" : np.mean,
        "height" : np.std
    }
)
```

```python
# función lambda
double_group.aggregate(
    {
        "Age" : np.mean,
        "height" : lambda h:(np.mean(h))/np.std(h)        
    }
)
```

```python
# Aplicamos a todos
double_group.aggregate([np.sum, np.mean, np.std])
```

```python
double_group.aggregate([lambda x: np.mean(x) / np.std(x)])
```

 ### Filtrado de datos

```python
# Filtramos por edad > 2400
double_group["Age"].filter(lambda x : x.sum() > 2400)
```

### Transformación de variables

```python
zscore = lambda x : (x - x.mean()) / x.std()
```

```python
# Se normalizan los grupos y se genera una campana de Gauss
z_group = double_group.transform(zscore)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(z_group["Age"])
```
