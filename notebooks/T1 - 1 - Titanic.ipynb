# Cargar datos a través de la función open

```python
rutabase = "C:/Users/Usuario/Documents/GitHub/cursoML/datasets"
rutadatos = "customer-churn-model/Customer Churn Model.txt"
rutacolumnas = "customer-churn-model/Customer Churn Columns.csv"
```

```python
# open puede ser lectura (r), escritura (w), lectura escritura (rw)
data_open = open(rutabase + "/" + rutadatos, 'r')
# leerLinea, eliminar espacion y separa por coma generando un array
cols = data_open.readline().strip().split(",")
# cantidad de columnas
cantidad_cols = len(cols)
```

```python
contador = 0
# generamos un diccionario vacio
main_dict = {}
# recorremo el array cols
for col in cols:
    # Crea la llave con la columna y el valor es un conjunto de elementos
    # por ahora vacio
    main_dict[col] = []
```

```python
# Recorremos cada una de las filas
for line in data_open:
    # generamos un array con cada elemento de cada fila
    valores = line.strip().split(",")
    # recorremos cols
    for i in range(len(cols)):
        # llenamos cada llave con sus respectivos valores
        main_dict[cols[i]].append(valores[i])
    contador += 1

print("El data set tiene %d filas y %d columnas"%(contador, cantidad_cols))
```

```python
dataframe_open = pd.DataFrame(main_dict)
```
