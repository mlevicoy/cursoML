### Simulación de Monte-Carlo

**Pasos**:

**1.** Generar dos números aleatorios uniformes entre x e y, entre 0 y 1 en total 1000 veces.

**2.** Calcular x * x + y * y

* Si el valor es menor o igual a 1, estamos dentro de círculo.
    
* Si el valor es mayor a 1, estamos fuera del círculo.

**3.** Calculamos el número total de veces que están dentro del círculo y lo dividimos entre el número total de intentos para obtener una aproximación de la probabilidad de caer dentro del círculo.

**4.** Usar dicha probabilidad para aproximar el valor de PI.

**5.** Repetir el experimento un número suficiente de veces (por ejemplo 100), para obtener diferentes aproximaciones de PI.

**6.** Calcular el promedio de los 1000 experimientos anteriores para dar un valor final de PI.

```python
def pi_montecarlo(n, n_exp):
    pi_avg = 0
    pi_value_list = []
    # Repetición de número de experimentos
    for i in range(n_exp):
        value = 0
        x = np.random.uniform(0,1,n).tolist()
        y = np.random.uniform(0,1,n).tolist()
        for j in range(n):
            z = np.sqrt(x[j] * x[j] + y[j] * y[j])
            if z <= 1:
                value += 1
        float_value = float(value)
        pi_value = float_value * 4 / n
        pi_value_list.append(pi_value)
        pi_avg += pi_value
    
    pi = pi_avg/n_exp
    
    print(pi)
    fig = plt.plot(pi_value_list)
    return (pi, fig)   
```

```python
pi_montecarlo(10000, 215)
```
